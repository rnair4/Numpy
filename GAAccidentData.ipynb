```python
import numpy as np
import pandas as pd
import requests
import lxml
import html5lib
from bs4 import BeautifulSoup
from pattern import web
url="http://www.gahighwaysafety.org/research/ga-crashes/injuries/fatalities/"
url
```

```python
website_html = requests.get(url).text
soupdata = BeautifulSoup(website_html, "html5lib")
table = soupdata.find("table", { "border" : "1" })
```

```python
gadata=[]
for record in table.findAll('td'):
        #goes through each td row in the table 
        gadata.append(record.text)
gadata[5]
```

```python
len(gadata)

type (gadata[0])
float(gadata[12].replace(',',''))

dic4 = {"Core Outcome Measure": [], "2007":[], "2008":[], "2009":[], "2010":[],"2011":[], "2011":[], "2012":[], "2013":[],"2016 Est":[]}

for i in range(11,209,9):
    dic4['Core Outcome Measure'].append(gadata[i].encode('ascii','ignore').replace('\n',' ').replace("\t",''))
    
    dic4['2007'].append(float(gadata[i+1].replace(',','')))
    dic4['2008'].append(float(gadata[i+1].replace(',','')))
    dic4['2009'].append(float(gadata[i+2].replace(',','')))
    dic4['2010'].append(float(gadata[i+3].replace(',','')))
    dic4['2011'].append(float(gadata[i+4].replace(',','')))
    dic4['2012'].append(float(gadata[i+5].replace(',','')))
    dic4['2013'].append(float(gadata[i+6].replace(',','')))
    dic4['2016 Est'].append(float(gadata[i+7].replace(',','')))
    
dic4
```
